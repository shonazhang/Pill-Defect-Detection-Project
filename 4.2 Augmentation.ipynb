```python
from matplotlib import image
from matplotlib import pyplot
# load image as pixel array
image = image.imread(path)
# summarize shape of the pixel array
print(image.dtype)
print(image.shape)
# display the array of pixels as an image
pyplot.imshow(image)
pyplot.show()
```

```python
#Horizontally flipped
hflipped_image= np.fliplr(image) #fliplr reverse the order of columns of pixels in matrix

#Vertically flipped
vflipped_image= np.flipud(image) #flipud reverse the order of rows of pixels in matrix

fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(10,16))
ax[0].imshow(image)
ax[0].set_title("Original Image", size=15)
ax[1].imshow(hflipped_image)
ax[1].set_title("Horizontally flipped", size=15)
ax[2].imshow(vflipped_image)
ax[2].set_title("Vertically flipped", size=15);
```

```python
import random
import pylab as pl 
def randRange(a, b):
    '''
    a utility function to generate random float values in desired range
    '''
    return pl.rand() * (b - a) + a
def randomCrop(im):
    '''
    croping the image in the center from a random margin from the borders
    '''
    margin = 1/3.5
    start = [int(randRange(0, im.shape[0] * margin)),
             int(randRange(0, im.shape[1] * margin))]
    end = [int(randRange(im.shape[0] * (1-margin), im.shape[0])), 
           int(randRange(im.shape[1] * (1-margin), im.shape[1]))]
    cropped_image = (im[start[0]:end[0], start[1]:end[1]])
    return cropped_image
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(8,10))
ax[0].imshow(image)
ax[0].set_title("Original Image", size=15)
ax[1].imshow(randomCrop(image))
ax[1].set_title("Cropped", size=15)
```

```python
image_bright = tf.image.adjust_gamma(image, gamma=0.5,gain=1)
image_dark = tf.image.adjust_gamma(image, gamma=2,gain=1)
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(10,16))
ax[0].imshow(image)
ax[0].set_title("Original Image", size=15)
ax[1].imshow(image_bright)
ax[1].set_title("Brightened Image", size=15)
ax[2].imshow(image_dark)
ax[2].set_title("Darkened Image", size=15)
```
